```python
import pandas as pd
import numpy as np
import os
```

```python
path = "./wlasl_csv"
wlasl_id_list = os.listdir(path)
```

```python
def csv_to_parquet(csv_df, video_id):
    csv_df['row_id'] = csv_df['frame'].astype('str')+'-'+csv_df['type']+'-'+csv_df['landmark_index'].astype('str')
    csv_df.drop(['Unnamed: 0', 'height', 'width'], axis=1, inplace=True)
    csv_df.to_parquet(path=f'./wlasl_parquet/{video_id}.parquet')
```

```python
for wlasl_id in wlasl_id_list:
    video_id = wlasl_id.split(".")[0]
    #print(video_id)
    csv_df = pd.read_csv(f'./wlasl_csv/{wlasl_id}')
    csv_to_parquet(csv_df, video_id)
```

```python
pd.read_parquet('./wlasl_parquet/01420.parquet')
```

```python
import json
with open('./wlasl_gloss.json') as f:
    wlasl_gloss = json.load(f)
```

```python
wlasl_gloss
```

```python
wlasl_data = []
for wlasl_id in wlasl_id_list:
    video_id = wlasl_id.split(".")[0]
    wlasl_data.append(video_id)
```

```python
wlasl_data
```

```python
train_dict = {}
train_dict['path'] = []
train_dict['sequence_id'] = []
train_dict['sign'] = []
for item in wlasl_data:
    p = f'wlasl_parquet/{item}.parquet'
    sq = item
    for elem in wlasl_gloss:
        #print(elem)
        for k,v in elem.items():
            #print(k)
            #print(v)
            #print(item)
            if item in v:
                sg = k
    train_dict['path'].append(p)
    train_dict['sequence_id'].append(sq)
    train_dict['sign'].append(sg)
```

```python
train_df = pd.DataFrame(train_dict, dtype=str)
```
